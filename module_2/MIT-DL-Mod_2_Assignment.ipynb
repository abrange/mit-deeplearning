## Prepare Data

```python
# These are the packages required for this assignment
import pandas as pd
import numpy as np

# Use Pandas to read the csv file into a dataframe.
# Note that the delimiter in this csv is the semicolon ";" instead of a ,
df_red = pd.read_csv('winequality-red.csv',delimiter=";")

# Because we are performing a classification task, we will assign all red wine a label of 1

df_red["color"] = 1

# The method .head() is super useful for seeing a preview of our data!
df_red.head()
```

```python
df_white = pd.read_csv('winequality-white.csv',delimiter=";")
df_white["color"] = 0  #assign white wine the label 0
df_white.head()
```

```python
# Now we combine our two dataframes
df = pd.concat([df_red, df_white])

# And shuffle them in place to mix the red and white wine data together
df = df.sample(frac=1).reset_index(drop=True)

SEED_ =  42
# Define the split ratio
test_size = 0.10
# For reproducibility
np.random.seed(SEED_)

# Sample 20% of the data for the test set
# frac=test_size specifies the fraction of rows to return
# random_state ensures reproducibility
test_df = df.sample(frac=test_size, random_state=SEED_)

# The remaining data will be the training set
# We drop the indices that are in the test set from the original DataFrame
train_df = df.drop(test_df.index)

print("Train DataFrame shape:", train_df.shape)
print("Test DataFrame shape:", test_df.shape)
print("\n")

# Separate in Train and Test Set. Train: 80% and Test 20%
train_df.head()
```

```python
# We choose three attributes of the wine to perform our prediction on
input_columns = ["citric acid", "residual sugar", "total sulfur dioxide"]
output_columns = ["color"]

# We extract the relevant features into our X and Y numpy arrays
X = train_df[input_columns].to_numpy()
Y = train_df[output_columns].to_numpy()
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
in_features = X.shape[1]
```

### Define Model

```python
import random
import math

class MyWineModel():
    def __init__(self, num_features, activation_function, do_print=False):
        self.w_0 = random.gauss(0,1)
        self.w = [random.gauss(0,1) for _ in range(num_features)]
        self.activation = activation_function
        self.do_print = do_print
    
    def forward(self, x):
        # Perform the dot product on the input x and the learned weights w
        z = 0
        for feature, weight in zip(x, self.w):
            z += feature * weight
        z += self.w_0  #Add the bias term
    
        # Apply the activation function, and return
        a = self.activation(z) #Example: sigmoid(z)
        return a

    # Training process, using negative log likelihood (NLL) loss --
    # appropriate for classification problems
    def train_model_NLL_loss(
            self,
            input_data,
            output_data,
            learning_rate,
            num_epochs):
        
        non_zero_tolerance = 1e-8 # add this to log calculations to ensure we don't take the log of 0
        for epoch in range(num_epochs):
            total_loss = 0 #keep track of total loss across the data set
            for x, y in zip(input_data, output_data):
                y_predicted = self.forward(x)
                nll_loss = -(y * math.log(y_predicted + non_zero_tolerance) + (1-y) * math.log(1-y_predicted + non_zero_tolerance))
    
                total_loss += nll_loss
                if self.do_print: print("x:", x, "y:", y, "nll_loss:", nll_loss)
                if self.do_print: print(" old weights:", self.w, self.w_0)
    
                # update bias coefficent using gradient w.r.t w_0
                self.w_0 -= learning_rate * (y_predicted - y)
    
                # update other model coefficients using gradient w.r.t that coeff
                for j, x_j in enumerate(x):
                    self.w[j] -= learning_rate * (y_predicted - y) * x_j
    
                if self.do_print: print(" new weights:", self.w,self.w_0)
    
            report_every = max(1, num_epochs // 10)
            if epoch % report_every == 0: #every few epochs, report on progress
                print("w_0:", self.w_0, "w", self.w)
                print("epoch", epoch, "has total loss", total_loss)

        return self.w, self.w_0
    
    # sigmoid activation function; squashes a real value z to between 0 and 1
    @staticmethod 
    def sigmoid(z):
        non_zero_tolerance = 1e-8 # add this to divisions to ensure we don't divide by 0
        return 1 / (1 + np.exp(-z) + non_zero_tolerance)

    # simple linear activation function: it just returns identity
    # AB Note: An identity function, is a function that returns the same input without modifing it.
    # example: f(x) = x
    @staticmethod
    def linear(z):
        return z
        
        
```

```python
#test model
model = MyWineModel(3, MyWineModel.sigmoid)
model.train_model_NLL_loss(X, Y, 0.00005, 500)
```
