### Define Model

```python
import random
import math

class MyWineModel():
    def __init__(self, num_features, activation_function, do_print=False):
        self.w_0 = random.gauss(0,1)
        self.w = [random.gauss(0,1) for _ in range(num_features)]
        self.activation = activation_function
        self.do_print = do_print
    
    def single_neuron_classification_model(self, x):
        # Perform the dot product on the input x and the learned weights w
        z = 0
        for feature, weight in zip(x, self.w):
            z += feature * weight
        z += self.w_0  #Add the bias term
    
        # Apply the activation function, and return
        a = self.activation(z) #Example: sigmoid(z)
        return a

    # Training process, using negative log likelihood (NLL) loss --
    # appropriate for classification problems
    def train_model_NLL_loss(
            self,
            input_data,
            output_data,
            learning_rate,
            num_epochs):
        
        non_zero_tolerance = 1e-8 # add this to log calculations to ensure we don't take the log of 0
        for epoch in range(num_epochs):
            total_loss = 0 #keep track of total loss across the data set
            for x, y in zip(input_data, output_data):
                y_predicted = self.single_neuron_classification_model(x)
                nll_loss = -(y * math.log(y_predicted + non_zero_tolerance) + (1-y) * math.log(1-y_predicted + non_zero_tolerance))
    
                total_loss += nll_loss
                if self.do_print: print("x:", x, "y:", y, "nll_loss:", nll_loss)
                if self.do_print: print(" old weights:", self.w, self.w_0)
    
                # update bias coefficent using gradient w.r.t w_0
                self.w_0 -= learning_rate * (y_predicted - y)
    
                # update other model coefficients using gradient w.r.t that coeff
                for j, x_j in enumerate(x):
                    self.w[j] -= learning_rate * (y_predicted - y) * x_j
    
                if self.do_print: print(" new weights:", self.w,self.w_0)
    
            report_every = max(1, num_epochs // 10)
            if epoch % report_every == 0: #every few epochs, report on progress
                print("w_0:", self.w_0, "w", self.w)
                print("epoch", epoch, "has total loss", total_loss)

        return self.w, self.w_0
    
    # sigmoid activation function; squashes a real value z to between 0 and 1
    @staticmethod 
    def sigmoid(z):
        non_zero_tolerance = 1e-8 # add this to divisions to ensure we don't divide by 0
        return 1 / (1 + math.exp(-z) + non_zero_tolerance)

    # simple linear activation function: it just returns identity
    # AB Note: An identity function, is a function that returns the same input without modifing it.
    # example: f(x) = x
    @staticmethod
    def linear(z):
        return z
    
    def forward(self, x):
        pass
        
        
```
